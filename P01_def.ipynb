## 1a. Hacer un programa que cargue una imagen y que obtenga y muestre el valor RGB de cada píxel de la imagen sobre el que esté el cursor del ratón.  

```python
import cv2 as cv
import numpy as np

def mouse(event, x, y, flags, param):
    bgr_pixel = param[y, x]
    # Convertir BGR a RGB (opcional pero recomendado para mostrar en RGB)
    rgb_pixel = bgr_pixel[[2, 1, 0]] # Invertir el orden para obtener RGB
    text = f'({x}, {y}), RGB: {rgb_pixel}'
    print(text)

    temp = img.copy()
    cv.putText(temp, text, (10, 30), cv.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)

    cv.imshow('EII', temp)

img = cv.imread('images/eii.png')

cv.imshow('EII', img)

cv.setMouseCallback('EII', mouse, img)

cv.waitKey(0)
cv.destroyAllWindows()
```

## 1b. Hacer un programa para dibujar las siguientes primitivas con el ratón 
- Líneas 
- Rectángulos 
- Círculos 
- Especificar el color en RGB de las primitivas 
- Definir el grosor de las líneas 

```python
import cv2 as cv
import numpy as np

def nothing(x):
    pass

cv.namedWindow('Dibujo')

# Trackbars para color
cv.createTrackbar('R','Dibujo',0,255,nothing)
cv.createTrackbar('G','Dibujo',0,255,nothing)
cv.createTrackbar('B','Dibujo',0,255,nothing)

# Trackbar para seleccionar primitiva: 0 = Línea, 1 = Rectángulo, 2 = Círculo
cv.createTrackbar('Primitiva', 'Dibujo', 0, 2, nothing)

# Trackbar para grosor de línea (mínimo 1, máximo 20)
cv.createTrackbar('Grosor', 'Dibujo', 1, 20, nothing)

cv.createTrackbar('Relleno', 'Dibujo', 0, 1, nothing) 

print("0: Líneas // 1: Rectángulos // 2: Círculos")

drawing = False
ix, iy = -1, -1
img = np.ones((500, 700, 3), np.uint8) * 255

def mouse(event, x, y, flags, param):
    global ix, iy, drawing, shape, color, thickness, canvas

    if event == cv.EVENT_LBUTTONDOWN:
        drawing = True
        ix, iy = x, y

    elif event == cv.EVENT_MOUSEMOVE and drawing:
        temp = img.copy()
        if shape == 'line':
            cv.line(temp, (ix, iy), (x, y), color, thickness)
        elif shape == 'rect':
            cv.rectangle(temp, (ix, iy), (x, y), color, thickness)
        elif shape == 'circle':
            radius = int(np.sqrt((x - ix) ** 2 + (y - iy) ** 2))    
            cv.circle(temp, (ix, iy), radius, color, thickness)

        cv.imshow('Dibujo', temp)

    elif event == cv.EVENT_LBUTTONUP:
        drawing = False
        if shape == 'line':
            cv.line(img, (ix, iy), (x, y), color, thickness)
        elif shape == 'rect':
            cv.rectangle(img, (ix, iy), (x, y), color, thickness)
        elif shape == 'circle':
            radius = int(np.sqrt((x - ix) ** 2 + (y - iy) ** 2))    
            cv.circle(img, (ix, iy), radius, color, thickness)

        cv.imshow('Dibujo', img)

cv.setMouseCallback('Dibujo', mouse)

while True:

    # Leer la posición actual de la trackbar de primitiva
    primitiva = cv.getTrackbarPos('Primitiva', 'Dibujo')

    # Elegir la forma según el valor
    if primitiva == 0:
        shape = 'line'
    elif primitiva == 1:
        shape = 'rect'
    elif primitiva == 2:
        shape = 'circle'

    # Leer los valores de color RGB
    r = cv.getTrackbarPos('R', 'Dibujo')
    g = cv.getTrackbarPos('G', 'Dibujo')
    b = cv.getTrackbarPos('B', 'Dibujo')
    color = (b, g, r)

     # Leer el grosor
    thickness = cv.getTrackbarPos('Grosor', 'Dibujo')

    relleno = cv.getTrackbarPos('Relleno', 'Dibujo')

    thickness = thickness if relleno == 0 else -1

    #cv.imshow('Dibujo', img)

    # Mostrar en consola
    print(f'Primitiva: {shape} | Color (R,G,B): ({r}, {g}, {b}) | Grosor: {thickness}   ', end='\r')

    # Salir con ESC
    k = cv.waitKey(100)
    if k == 27:
        break

cv.destroyAllWindows()
```

```python
import cv2 as cv
import numpy as np
import datetime

def nothing(x):
    pass

cv.namedWindow('Dibujo')

cv.createTrackbar('R','Dibujo',0,255,nothing)
cv.createTrackbar('G','Dibujo',0,255,nothing)
cv.createTrackbar('B','Dibujo',0,255,nothing)

cv.createTrackbar('Primitiva', 'Dibujo', 0, 5, nothing)

cv.createTrackbar('Grosor', 'Dibujo', 1, 20, nothing)

cv.createTrackbar('Relleno', 'Dibujo', 0, 1, nothing) 

print("0: Líneas // 1: Rectángulos // 2: Círculos // 3: Elipses // 4: Polilíneas // 5: Polígonos")
print("0: Contorno // 1: Relleno")
print("ESC: Salir // 'c': Limpiar // 'g': Guardar imagen")

drawing = False
ix, iy = -1, -1
img = np.ones((500, 700, 3), np.uint8) * 255
puntos = []

height, width, _ = img.shape        
fps = 24                             
fourcc = cv.VideoWriter_fourcc(*'mp4v') 
filename = datetime.datetime.now().strftime("dibujo_%Y%m%d_%H%M%S.mp4")
out = cv.VideoWriter(filename, fourcc, fps, (width, height))

def mouse(event, x, y, flags, param):
    global ix, iy, drawing, shape, color, thickness, canvas

    if shape in ['line', 'rect', 'circle', 'ellipse']:
        if event == cv.EVENT_LBUTTONDOWN:
            drawing = True
            ix, iy = x, y

        elif event == cv.EVENT_MOUSEMOVE and drawing:
            temp = img.copy()
            if shape == 'line':
                cv.line(temp, (ix, iy), (x, y), color, thickness)
            elif shape == 'rect':
                cv.rectangle(temp, (ix, iy), (x, y), color, thickness_actual)
            elif shape == 'circle':
                radius = int(np.sqrt((x - ix) ** 2 + (y - iy) ** 2))    
                cv.circle(temp, (ix, iy), radius, color, thickness_actual)
            elif shape == 'ellipse':
                axes = (abs(x - ix), abs(y - iy))
                cv.ellipse(temp, (ix, iy), axes, 0, 0, 360, color, thickness_actual)

            cv.imshow('Dibujo', temp)
            out.write(temp)

        elif event == cv.EVENT_LBUTTONUP:
            drawing = False
            if shape == 'line':
                cv.line(img, (ix, iy), (x, y), color, thickness)
            elif shape == 'rect':
                cv.rectangle(img, (ix, iy), (x, y), color, thickness_actual)
            elif shape == 'circle':
                radius = int(np.sqrt((x - ix) ** 2 + (y - iy) ** 2))    
                cv.circle(img, (ix, iy), radius, color, thickness_actual)
            elif shape == 'ellipse':
                axes = (abs(x - ix), abs(y - iy))
                cv.ellipse(img, (ix, iy), axes, 0, 0, 360, color, thickness_actual)

            cv.imshow('Dibujo', img)
            out.write(img)
    
    elif shape in ['polyline', 'polygon']:
        if event == cv.EVENT_LBUTTONDOWN:
            puntos.append((x, y))
            temp = img.copy()  
            for i in range(1, len(puntos)):
                cv.line(temp, puntos[i-1], puntos[i], color, thickness)
            cv.imshow('Dibujo', temp)
            out.write(temp)

        elif event == cv.EVENT_RBUTTONDOWN and len(puntos) > 1:

            if shape == 'polygon':
                puntos_array = np.array(puntos, dtype=np.int32).reshape((-1, 1, 2))
                
                if relleno == 1:
                    cv.fillPoly(img, [puntos_array], color)

                cv.polylines(img, [puntos_array], isClosed=True, color=color, thickness=thickness)
                cv.imshow('Dibujo', img)
                out.write(img)

            elif shape == 'polyline':
                puntos_polylines = np.array(puntos, dtype=np.int32).reshape((-1, 1, 2))

                cv.polylines(img, [puntos_polylines], isClosed=False, color=color, thickness=thickness)
                cv.imshow('Dibujo', img)
                out.write(img)

            puntos.clear()  
        
cv.setMouseCallback('Dibujo', mouse)

while True:

    primitiva = cv.getTrackbarPos('Primitiva', 'Dibujo')

    if primitiva == 0:
        shape = 'line'
    elif primitiva == 1:
        shape = 'rect'
    elif primitiva == 2:
        shape = 'circle'
    elif primitiva == 3:
        shape = 'ellipse'
    elif primitiva == 4:
        shape = 'polyline'
    elif primitiva == 5:
        shape = 'polygon'

    r = cv.getTrackbarPos('R', 'Dibujo')
    g = cv.getTrackbarPos('G', 'Dibujo')
    b = cv.getTrackbarPos('B', 'Dibujo')
    color = (b, g, r)

    thickness = cv.getTrackbarPos('Grosor', 'Dibujo')

    relleno = cv.getTrackbarPos('Relleno', 'Dibujo')

    thickness_actual = thickness if relleno == 0 else -1

    print(f'Primitiva: {shape} | Color (R,G,B): ({r}, {g}, {b}) | Grosor: {thickness} | Relleno: {relleno}   ', end='\r')

    k = cv.waitKey(100)
    if k == 27:
        break

    elif k == ord('c'):  
        img[:] = 255
        puntos.clear()

    elif k == ord('g'): 
        filename = datetime.datetime.now().strftime("dibujo_%Y%m%d_%H%M%S.png")
        cv.imwrite(filename, img)
    

out.release()
cv.destroyAllWindows()
```
